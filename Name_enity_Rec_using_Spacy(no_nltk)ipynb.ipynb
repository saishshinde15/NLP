```python
import spacy
from spacy import displacy
from collections import Counter
import en_core_web_sm
nlp = en_core_web_sm.load()
```

```python
from bs4 import BeautifulSoup
import requests
import re
def url_to_string(url):
    res = requests.get(url)
    html = res.text
    soup = BeautifulSoup(html, 'html5lib')
    for script in soup(["script", "style", 'aside']):
        script.extract()
    return " ".join(re.split(r'[\n\t]+', soup.get_text()))
ny_bb = url_to_string('https://appleinsider.com/inside/a12z-bionic')
article = nlp(ny_bb)
len(article.ents)
```

```python
labels = [x.label_ for x in article.ents]
Counter(labels)
```

```python
items = [x.text for x in article.ents]
Counter(items).most_common(3)
```

```python
sentences = [x for x in article.sents]
print(sentences[10])# randomly selecting a sentenece (10th line)
```

```python
displacy.render(nlp(str(sentences[20])), jupyter=True, style='ent')
```

```python
displacy.render(nlp(str(sentences[10])), style='dep', jupyter = True, options = {'distance': 120})
```

```python
[(x.orth_,x.pos_, x.lemma_) for x in [y
                                      for y
                                      in nlp(str(sentences[10]))
                                      if not y.is_stop and y.pos_ != 'PUNCT']]#xtract part-of-speech and lemmatize this sentence.
```

```python
dict([(str(x), x.label_) for x in nlp(str(sentences[10])).ents])
```
